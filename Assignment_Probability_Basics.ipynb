## QUESTION 1
**A die is rolled. What is the probability of getting:**  
(a) An even number  
(b) A number greater than 4

**Answer:**  
(a) **1/2** (Even numbers: 2, 4, 6 → 3 favourable out of 6)  
(b) **1/3** (Numbers > 4: 5, 6 → 2 favourable out of 6)

## QUESTION 2
In a class of 50 students:  
- 20 like Mathematics (M)  
- 15 like Science (S)  
- 5 like both subjects  

What is the probability that a student chosen at random likes Mathematics or Science?

**Answer:**  
**P(M or S) = 3/5 or 0.6**  
Calculation: (20 + 15 - 5)/50 = 30/50 = 3/5

---

## QUESTION 3
A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?

**Answer:**  
**1/2**  
After drawing one red ball: 2 red and 2 blue left (total 4).  
Probability = 2/4 = 1/2

## QUESTION 4
The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?

**Answer:**  
**Stratified Sampling**  
Why? It divides the population into strata (boys & girls) and ensures proportional or equal representation from each group. Simple Random Sampling may not guarantee this balance.

## QUESTION 5
The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm.  
Find the sampling error.

**Answer:**  
**Sampling error = 2 cm**  
(160 cm – 158 cm)

---

## QUESTION 6
The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?

**Answer:**  
**SEM = ₹500**  
Formula: σ / √n = 5000 / √100 = 5000 / 10 = 500

## QUESTION 7
In a group of 100 students:  
- 40 like Cricket (C)  
- 30 like Football (F)  
- 10 like both Cricket and Football  

Find the probability that a student likes at least one sport.

**Answer:**  
**0.6 or 3/5**  
P(C or F) = (40 + 30 - 10)/100 = 60/100 = 0.6

## QUESTION 8
From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?

**Answer:**  
**1/221**  
P(both aces) = (4/52) × (3/51) = 12 / (52 × 51) = 1/221

## QUESTION 10
Differentiate between discrete and continuous random variables with examples.

**Answer:**  
**Discrete Random Variable:**  
Takes countable (finite or countably infinite) values.  
Example: Number of heads in 5 coin tosses (0,1,2,3,4,5), Number of students absent in a class.

**Continuous Random Variable:**  
Can take any value in a continuous range (infinite possible values).  
Example: Height of students (any value between 150 cm and 180 cm), Time taken to complete a test.
